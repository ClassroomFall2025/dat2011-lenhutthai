```python
# Tạo hàm
def nhap_nhanvien():
    print("\nChức năng 1: Nhập danh sách nhân viên từ bàn phím")

def xuat_nhanvien():
    print("\nChức năng 2: Đọc thông tin nhân viên từ file và xuất danh sách nhân viên ra màn hình")

def tim_nhanvien():
    print("\nChức năng 3: Tìm và hiển thị nhân viên theo mã nhập từ bàn phím")

def xoa_nhanvien():
    print("\nChức năng 4: Xóa nhân viên theo mã nhập từ bàn phím")

def capnhat_nhanvien():
    print("\nChức năng 5: Cập nhật thông tin nhân viên theo mã nhập từ bàn phím")

def tim_theoluong():
    print("\nChức năng 6: Tìm các nhân viên theo khoảng lương nhập")

def sapxep_hoten():
    print("\nChức năng 7: Sắp xếp nhân viên theo họ tên")

def sapxep_thunhap():
    print("\nChức năng 8: Sắp xếp nhân viên theo thu nhập")

def top5_thunhap():
    print("\nChức năng 9: Hiển thị 5 nhân viên có thu nhập cao nhất")

def menu():
    while True:
        print("\n====================== HỆ THỐNG QUẢN LÝ NHÂN VIÊN ======================")
        print("0. Thoát")
        print("1. Nhập danh sách nhân viên từ bàn phím")
        print("2. Đọc thông tin nhân viên từ file và xuất danh sách nhân viên ra màn hình")
        print("3. Tìm và hiển thị nhân viên theo mã nhập từ bàn phím")
        print("4. Xóa nhân viên theo mã nhập từ bàn phím")
        print("5. Cập nhật thông tin nhân viên theo mã nhập từ bàn phím")
        print("6. Tìm các nhân viên theo khoảng lương nhập")
        print("7. Sắp xếp nhân viên theo họ tên")
        print("8. Sắp xếp nhân viên theo thu nhập")
        print("9. Hiển thị 5 nhân viên có thu nhập cao nhất")
        print("==========================================================================")
        

        chon = input("Nhập lựa chọn của bạn: ")

        match chon:
            case "0":
                print("Thoát chương trình.")
                break
            case "1":
                nhap_nhanvien()
            case "2":
                xuat_nhanvien()
            case "3":
                tim_nhanvien()
            case "4":
                xoa_nhanvien()
            case "5":
                capnhat_nhanvien()
            case "6":
                tim_theoluong()
            case "7":
                sapxep_hoten()
            case "8":
                sapxep_thunhap()
            case "9":
                top5_thunhap()

# Gọi menu
menu()
```
